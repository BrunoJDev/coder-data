```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('https://raw.githubusercontent.com/BrunoJDev/coder-data/main/ign.csv')

# Información acerca de los datos
df.info()
df.describe()
df.head()

## Gráficos con Matplotlib

# Gráfico 1: Distribución de puntuaciones
plt.figure(figsize=(10, 6))
plt.hist(df['score'], bins=10, edgecolor='black')
plt.title('Distribución de Puntuaciones')
plt.xlabel('Puntuación')
plt.ylabel('Frecuencia')

# Gráfico 2: Puntuaciones por plataforma
plt.figure(figsize=(10, 6))
df.groupby('platform')['score'].mean().plot(kind='bar')
plt.title('Puntuaciones por Plataforma')
plt.xlabel('Plataforma')
plt.ylabel('Puntuación Media')

# Gráfico 3: Puntuaciones por género
plt.figure(figsize=(10, 6))
df.groupby('genre')['score'].mean().plot(kind='bar')
plt.title('Puntuaciones por Género')
plt.xlabel('Género')
plt.ylabel('Puntuación Media')

plt.tight_layout()
plt.show()

## Gráficos con Seaborn

# Gráfico 4: Puntuaciones por frase de puntuación
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='score_phrase', y='score')
plt.title('Puntuaciones por Frase de Puntuación')
plt.xlabel('Frase de Puntuación')
plt.ylabel('Puntuación')

# Gráfico 5: Puntuaciones por elección del editor
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='editors_choice', y='score')
plt.title('Puntuaciones por Elección del Editor')
plt.xlabel('Elección del Editor')
plt.ylabel('Puntuación Media')

# Gráfico 6: Puntuaciones por año de lanzamiento
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='release_year', y='score', marker='o')
plt.title('Puntuaciones por Año de Lanzamiento')
plt.xlabel('Año de Lanzamiento')
plt.ylabel('Puntuación')

plt.tight_layout()
plt.show()
```
