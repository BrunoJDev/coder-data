```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('https://raw.githubusercontent.com/BrunoJDev/coder-data/main/ign.csv')

# Informacin acerca de los datos
df.info()
df.describe()
df.head()

## Grficos con Matplotlib

# Grfico 1: Distribucin de puntuaciones
plt.figure(figsize=(10, 6))
plt.hist(df['score'], bins=10, edgecolor='black')
plt.title('Distribucin de Puntuaciones')
plt.xlabel('Puntuacin')
plt.ylabel('Frecuencia')

# Grfico 2: Puntuaciones por plataforma
plt.figure(figsize=(10, 6))
df.groupby('platform')['score'].mean().plot(kind='bar')
plt.title('Puntuaciones por Plataforma')
plt.xlabel('Plataforma')
plt.ylabel('Puntuacin Media')

# Grfico 3: Puntuaciones por gnero
plt.figure(figsize=(10, 6))
df.groupby('genre')['score'].mean().plot(kind='bar')
plt.title('Puntuaciones por Gnero')
plt.xlabel('Gnero')
plt.ylabel('Puntuacin Media')

plt.tight_layout()
plt.show()

## Grficos con Seaborn

# Grfico 4: Puntuaciones por frase de puntuacin
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='score_phrase', y='score')
plt.title('Puntuaciones por Frase de Puntuacin')
plt.xlabel('Frase de Puntuacin')
plt.ylabel('Puntuacin')

# Grfico 5: Puntuaciones por eleccin del editor
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='editors_choice', y='score')
plt.title('Puntuaciones por Eleccin del Editor')
plt.xlabel('Eleccin del Editor')
plt.ylabel('Puntuacin Media')

# Grfico 6: Puntuaciones por ao de lanzamiento
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='release_year', y='score', marker='o')
plt.title('Puntuaciones por Ao de Lanzamiento')
plt.xlabel('Ao de Lanzamiento')
plt.ylabel('Puntuacin')

plt.tight_layout()
plt.show()
```
